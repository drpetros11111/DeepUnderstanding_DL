```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# NEW!
import scipy.stats as stats
```

```python
# parameters
n1 = 30   # samples in dataset 1
n2 = 40   # ...and 2
mu1 = 1   # population mean in dataset 1
mu2 = 2   # population mean in dataset 2

# generate the data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)



# plot them
plt.plot(np.zeros(n1),data1,'ro',markerfacecolor='w',markersize=14)
plt.plot(np.ones(n2), data2,'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1','Group 2'])
plt.show()
```

```python
# t-test via stats package

# _ind = independent samples
t,p = stats.ttest_ind(data1,data2)
print(t)
print(p)
```

# Interpreting the Results of the t-test

    t-statistic: -3.2079668300064066

##t-statistic:

The t-statistic is negative, which suggests that the mean of data1 is lower than the mean of data2.

The magnitude of the t-statistic (3.21) indicates that the difference between the two group means is relatively large compared to the variability within the groups.

-------------
##p-value:

The p-value (0.002) is much smaller than the typical threshold of 0.05 (or even 0.01).

This means that the probability of observing such a large difference between the means of the two groups by random chance is very low.

--------------------
#Conclusion

You can reject the null hypothesis.

There is strong evidence to suggest that the two samples (data1 and data2) have significantly different means.

-------------------
#Summary
The result indicates that there is a statistically significant difference between the two groups.

The p-value is small enough to reject the null hypothesis, meaning the two samples do not come from populations with the same mean.

Specifically, data1 has a lower mean compared to data2.

# Two-sample t-test
This code performs an independent two-sample t-test using the scipy.stats package.

A t-test is a statistical test used to determine if there is a significant difference between the means of two groups.

---------------------

    t, p = stats.ttest_ind(data1, data2)

##stats.ttest_ind(data1, data2):

This function from the scipy.stats module calculates the t-statistic and the p-value for an independent two-sample t-test between data1 and data2.

---------------
#What the t-test does:

It compares the means of two independent groups (data1 and data2) to see if they are significantly different from each other.

##The null hypothesis (H₀) of the t-test is that the means of the two groups are equal.

The alternative hypothesis (H₁) is that the means of the two groups are different.

---------------------------
#Parameters:
##data1 and data2:

These are the two independent samples (sets of data) being compared.

Each sample represents a different group.

---------------------
#Returned Values:
##t:

This is the t-statistic. It measures the difference between the groups relative to the variability in the data.

A larger absolute value of t indicates that the group means are more different from each other.

Positive or negative values: The sign indicates the direction of the difference between the two means.

Larger magnitude (positive or negative) suggests a more significant difference between group means.

p: This is the p-value. It tells you the probability of observing a result as extreme as, or more extreme than, the one observed if the null hypothesis is true (i.e., if there really is no difference between the means of the two groups).

##Low p-value (typically < 0.05):

This suggests that the null hypothesis can be rejected, meaning the means of the two groups are significantly different.

##High p-value (≥ 0.05):

This suggests that there is no significant difference between the group means, and the observed difference could be due to random chance.

------------------------
#Example Output:

print(t)  # Prints the t-statistic value
print(p)  # Prints the p-value

If the output of p is less than 0.05 (common threshold for significance):

You can reject the null hypothesis, indicating that data1 and data2 likely have different means.

If the output of p is greater than 0.05:

You fail to reject the null hypothesis, meaning there is not enough evidence to conclude that the means of data1 and data2 are significantly different.

-------------------------
#Summary
t: Measures the size of the difference relative to the variation in the data.

p: Tells you whether the difference is statistically significant.

```python
# common way to show t-test results in a plot
fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':12}) # change the font size

plt.plot(0+np.random.randn(n1)/15, data1, 'ro',markerfacecolor='w',markersize=14)
plt.plot(1+np.random.randn(n2)/15, data2, 'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1','Group 2'])

# set the title to include the t-value and p-value
plt.title(f't = {t:.2f}, p={p:.3f}')

plt.show()
```
