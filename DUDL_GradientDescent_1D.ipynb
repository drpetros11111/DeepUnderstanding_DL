# Gradient descent in 1D

```
# function (as a function)
def fx(x):
  return 3*x**2 - 3*x + 4

# derivative function
def deriv(x):
  return 6*x - 3
```

# Calculating the Local Min
---------------------------
image.png

---------------------------------
Let's break it down step by step:

#1. Random Starting Point:

    localmin = np.random.choice(x, 1)
    print(localmin)

##
What it does:

np.random.choice(x, 1) randomly selects one value from the array x, which contains values between -2 and 2.

This value serves as the initial guess or starting point for gradient descent.
localmin: This will hold the current value of the guess (starting with a random value).

-------------------------
#2. Learning Parameters:

    learning_rate = .01
    training_epochs = 100

What they mean:
learning_rate: This controls how big each step is in the direction of the negative gradient (the step size). A smaller learning rate means smaller steps, and a larger one means bigger steps.
training_epochs: The number of iterations (or steps) for the gradient descent process. The code will update the guess for the local minimum 100 times.

-------------------
#3. Gradient Descent Loop:

    for i in range(training_epochs):
      grad = deriv(localmin)
      localmin = localmin - learning_rate * grad

##How it works:
The loop runs 100 iterations (defined by training_epochs).

grad = deriv(localmin): This calculates the gradient (or derivative) of the function fx(x) at the current point localmin.

The derivative tells us the slope of the function at the current point. If the slope is positive, it means the function is increasing, so we need to move left (downhill); if the slope is negative, we need to move right.

localmin = localmin - learning_rate * grad: This is the gradient descent step.

The point localmin is updated by subtracting learning_rate * grad. This means we move the current point in the direction of the steepest descent (downhill).

The learning rate controls how large the step is. If the learning rate is too small, it will take many steps to reach the minimum. If it’s too large, we might overshoot and miss the minimum.

-------------------------
#4. Result:

    localmin
After the loop finishes, localmin will contain the final value of the local minimum found through gradient descent.

This value is the point where the derivative becomes close to zero, indicating that we've reached a low point in the function.

-----------------------
#Conceptual Explanation:
Gradient Descent is an optimization algorithm that iteratively moves toward the local minimum of a function by updating the starting point in the opposite direction of the gradient.

In this case, the function fx(x) is quadratic, so gradient descent will eventually find the minimum value of this curve (which is a parabola).

----------------
#Visualization:
If you visualize fx(x) as a U-shaped curve (a parabola), gradient descent is like starting from a random point on the curve and walking downhill step by step until you reach the bottom (the minimum).

-----------------
#Key Points:
Initial Point: A random starting point is chosen.

Learning Rate: Controls how big the steps are.

Gradient: The slope of the function, which tells you the direction to move.

Update: You subtract a fraction of the gradient from the current point to move towards the minimum.

This process repeats until the function reaches a point where the slope is nearly zero (indicating a flat area, or minimum).

```
# plot the results
!pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np

# random starting point
localmin = np.random.choice(x,1)

# learning parameters
learning_rate = .01
training_epochs = 100

# run through training
for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad

plt.plot(x,fx(x), x,deriv(x))
plt.plot(localmin,deriv(localmin),'ro')
plt.plot(localmin,fx(localmin),'ro')

```

# Store the model parameters and outputs on each iteration

```
# random starting point
localmin = np.random.choice(x,1)

# learning parameters
learning_rate = .01
training_epochs = 100

# run through training and store all the results
modelparams = np.zeros((training_epochs,2))
for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad
  modelparams[i,0] = localmin
  modelparams[i,1] = grad
```

```
# plot the gradient over iterations

fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
  ax[i].plot(modelparams[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {localmin[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```
